```python
import cv2
import numpy as np

def cv_show_image(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def is_point_inside_rect(x, y, rect_x, rect_y, rect_width, rect_height):
    # 判斷點(x, y)是否在長方形內部
    if rect_x <= x <= rect_x + rect_width and rect_y <= y <= rect_y + rect_height:
        return True
    else:
        return False

def blob(img, rects):
    im = cv2.imread(img, cv2.IMREAD_COLOR)
    
    # 創建SimpleBlobDetector參數
    params = cv2.SimpleBlobDetector_Params()
    
    # 設置檢測器參數（參考你之前的設置）
    params.minThreshold = 10
    params.maxThreshold = 240
    
    params.filterByColor = True
    params.blobColor = 255
    
    params.filterByArea = True
    params.minArea = 100
    params.maxArea = 1050
    
    params.filterByCircularity = True
    params.minCircularity = 0.5
    params.maxCircularity = 1.0
    
    params.filterByConvexity = True
    params.minConvexity = 0.5
    params.maxConvexity = 1.0
    
    params.filterByInertia = True
    params.minInertiaRatio = 0.02
    
    # 創建檢測器
    detector = cv2.SimpleBlobDetector_create(params)
    
    # 檢測斑點
    keypoints = detector.detect(im)
    
    # 遍歷檢測到的每個斑點
    for keypoint in keypoints:
        x, y = int(keypoint.pt[0]), int(keypoint.pt[1])
        # 初始化標記變數為False
        is_inside_rect = False
        # 遍歷每個長方形，判斷斑點是否在長方形內部
        for rect in rects:
            rect_x, rect_y, rect_width, rect_height = rect
            if is_point_inside_rect(x, y, rect_x, rect_y, rect_width, rect_height):
                # 在長方形內部畫上紅色圓圈
                im = cv2.drawKeypoints(im, [keypoint], np.array([]), (0, 0, 255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
                # 將標記變數設定為True，表示該斑點至少在一個長方形內
                is_inside_rect = True
                break  # 如果在一個長方形內，則無需繼續檢查其他長方形
        # 如果標記變數為False，則該斑點不在任何長方形內，不進行標記
        
    # 顯示結果
    cv2.imwrite("result_frame_21.jpg", im)

# 讀取txt文件，獲取長方形位置信息
rects = []
with open("21.csv", "r", newline='') as file:
    for line in file:
        # 去除行末的逗號，並提取每個長方形的位置信息
        x, y, width, height= map(float, row[2:6])
        rects.append((x, y, width, height))
        #print(f"讀取到的長方形位置信息：x={x}, y={y}, width={width}, height={height}")
        print(rects)

'''# 找到所有矩形的最小面積和最大面積
min_area = min(rect[4] for rect in rects)
max_area = max(rect[4] for rect in rects)

# 創建新的SimpleBlobDetector參數
params = cv2.SimpleBlobDetector_Params()

# 將params.minArea和params.maxArea設定為讀取的area值加減50
params.minArea = max(min_area - 50, 1)  # 最小面積不能小於1
params.maxArea = max_area + 50'''

# 檢測並繪製紅色圓圈
blob("frame_21.jpg", rects)
```
